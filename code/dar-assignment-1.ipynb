```r
filtered_diamonds <- diamonds %>%
  filter(price > 10000 & carat > 1)
```

```r
sorted_diamonds <- filtered_diamonds %>%
  arrange(desc(price))
```

```r
summary_stats <- sorted_diamonds %>%
  summarise(avg_price = mean(price), avg_carat = mean(carat))
```

```r
print(summary_stats)
```
