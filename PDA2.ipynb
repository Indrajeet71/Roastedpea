```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('Lab_2_BEML_gain.csv')
```

```python
data
```

```python
data.info()
```

```python
data['daily_return'] = data['Close'].pct_change()
```

```python
data
```

```python
prob_gain_3_percent = np.sum(data['daily_return'] >= 0.03) / len(data)
prob_loss_2_percent = np.sum(data['daily_return'] <= -0.02) / len(data)
```

```python
print(f'Probability of making a daily return of 3% or more: {prob_gain_3_percent}')
print(f'Probability of making a loss of 2% or more: {prob_loss_2_percent}')
```
